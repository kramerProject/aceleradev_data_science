# Análise Exploratória

```python
aux = pd.DataFrame({'variables': df_merged.columns,
                    'tipos': df_merged.dtypes,
                    'percentual_faltante': df_merged.isna().sum() / df_merged.shape[0]})
data_missing_higher_50 = aux[aux['percentual_faltante']>0.5]
data_missing_lower_equal_50 = aux[aux['percentual_faltante']<=0.5]

variables_missing_higher_50 = [item for item in data_missing_higher_50['variables']]
variables_missing_lower_equal_50 = [item for item in data_missing_lower_equal_50['variables']]
```

## Analisando dados faltantes

```python
# returns for each column their values and quatity. And percentage
def val_counts():
    columns = variables_missing_lower_equal_50
    for column in columns:
        print(f'-------> Coluna: {column}')
        print(df[column].value_counts())
        obj = {
            "total": len(df[column]),
            "percentages": (df[column].value_counts()/df[column].count())*100
        }
        print(obj)
    
# val_counts()
```

### Das 185 colunas identificamos 89 variaveis com menos de 50% dos dados faltantes

```python
len(variables_missing_lower_equal_50)
```

```python

new_data = df_merged[variables_missing_lower_equal_50]
```

# Variaveis por tipo em ordem decrescente em % de dados faltantes

```python
pd.set_option("max_rows", None)
aux = pd.DataFrame({'variables': new_data.columns,
                    'var_types': new_data.dtypes,
                    'missing_percentages': new_data.isna().sum() / new_data.shape[0]})
pivot = pd.pivot_table(aux, index=['var_types', 'variables'])
pivot.sort_values(by=['var_types','missing_percentages'],  ascending=False)
```
